```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = np.load('geo_spatial_datav2.npy')
df = pd.DataFrame(data)
```

```python
df_transf = pd.DataFrame(columns=['latitude', 'longitude', 'magnitude'])
df_transf
```

```python
for lat, long in df.iterrows():
    lat_all = [lat]*len(df.columns)
    df_temp = pd.DataFrame({'latitude': lat_all, 
                        'longitude': long.index.values.tolist(), 
                        'magnitude': long.values.tolist()
                        })
    df_transf = pd.concat([df_transf, df_temp])
```

```python
df_transf['longitude'] = df_transf['longitude']/100 + 15.0
df_transf['latitude'] = ((df_transf['latitude']/100)*-1) - 22.0
```

```python
df_transf.to_csv('data_transformed.csv')
```
